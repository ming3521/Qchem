## 使用 chemistry 模块获取分子哈密顿量、核排斥能与基态能量等信息

### 导入必要的库

```python
from chemistry.molecule import Molecule
from openfermion.linalg import get_sparse_operator, get_ground_state
```

### 加载 molecule 模块
在 symbols 处输入分子类型，在 geometry 处输入分子坐标，通过 basis 选择想使用的基组，在 multiplicity 处设置多重态，在 charge 处输入电荷数。

```python
molecule = Molecule(
    symbols = ["N","H","H","H"], 
    geometry = [[0.0, 0.0, 0.1173], [0.0, 0.9397, -0.2641],[0.8138, -0.4699, -0.2641],[-0.8138, -0.4699, -0.2641]], 
    basis = 'sto-3g', 
    multiplicity = 1, 
    charge = 0
)
molecule.run()
```

### 活性空间近似
设置活性空间，选择映射类型为 Jordan-Wigner 映射

```python
molecule.set_active_space(2, 2)
hamiltonian = molecule.qubit_hamiltonian("jw")
hf_state = molecule.hf_state("jw")

for i, op in enumerate(hamiltonian.terms):
    print(op, hamiltonian.terms[op])
    if i > 100:
        break
print(hf_state)
```

### 输出核排斥能等能量信息

```python
hamiltonian_sparse = get_sparse_operator(hamiltonian)
ground_energy, ground_state_vector = get_ground_state(hamiltonian_sparse)

print(f"核排斥能: {molecule.data.nuclear_repulsion:.12f} Hartree")
print(f"Hartree-Fock 能量: {molecule.data.hf_energy:.12f} Hartree")
print(f"冻结轨道后基态能量: {ground_energy:.12f} Hartree")
```
